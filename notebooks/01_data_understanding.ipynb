# RQ1: Sample Size Justification

**Research Question 1**

> Which network telemetry features contribute most to accurately identifying network flow anomalies?

Before developing machine learning models, it is important to verify that the available dataset is sufficiently large to support reliable statistical analysis and model training.

Although the UNSW-NB15 dataset already provides predefined training and testing partitions, this study formally evaluates sample adequacy using two complementary approaches:

1. **Green's Rule** – a heuristic for estimating the minimum recommended sample size for multivariable analyses based on the number of predictor variables.
3. **Events Per Variable (EPV)** – a widely accepted criterion for classification models that measures the number of positive events available for each predictor variable.

Together, these measures provide statistical evidence that the dataset is sufficiently large for feature importance analysis and supervised machine learning.

```python
import pandas as pd

# Load the training dataset
df = pd.read_csv("/Users/shreekkanthkalyanasundaram/AIML/capstone_Q640/UNSW_NB15_training-set.csv")

# Predictor variables (42 features)
X = df.drop(columns=["id", "attack_cat", "label"])

# Target variable (Binary classification)
y = df["label"]

# Sample statistics
n_samples = len(df)
n_predictors = X.shape[1]
n_events = (y == 1).sum()      # Anomalous network flows
n_non_events = (y == 0).sum()  # Normal network flows

# Green's Rule
green_rule = 50 + 8 * n_predictors

# Events Per Variable (EPV)
epv = n_events / n_predictors

print(f"Total Samples              : {n_samples}")
print(f"Predictor Variables        : {n_predictors}")
print(f"Positive Events (Anomaly)  : {n_events}")
print(f"Negative Events (Normal)   : {n_non_events}")
print(f"Green's Rule Minimum       : {green_rule}")
print(f"Events Per Variable (EPV)  : {epv:.2f}")
```

```python
# ============================================================
# RESEARCH QUESTION 1
# Sample Size Justification
# ============================================================

"""
Research Question 1

Which network telemetry features contribute most to accurately
identifying operational anomalies?

Sample Size Methods:
1. Green's Rule
2. Events Per Variable (EPV)
"""

import pandas as pd

# ------------------------------------------------------------
# Load Training Dataset
# ------------------------------------------------------------

df = pd.read_csv("/Users/shreekkanthkalyanasundaram/AIML/capstone_Q640/UNSW_NB15_training-set.csv")

# ------------------------------------------------------------
# Predictor Variables and Target Variable
# ------------------------------------------------------------

X = df.drop(columns=["id", "attack_cat", "label"])
y = df["label"]

# ------------------------------------------------------------
# Dataset Statistics
# ------------------------------------------------------------

n_samples = len(df)
n_predictors = X.shape[1]

n_events = (y == 1).sum()
n_non_events = (y == 0).sum()

# ------------------------------------------------------------
# Green's Rule
# ------------------------------------------------------------

green_rule = 50 + 8 * n_predictors

# ------------------------------------------------------------
# Events Per Variable (EPV)
# ------------------------------------------------------------

epv = n_events / n_predictors

# ------------------------------------------------------------
# Results
# ------------------------------------------------------------

print("=" * 70)
print("RESEARCH QUESTION 1")
print("=" * 70)

print("\nResearch Question")
print("-" * 70)
print("Which network telemetry features contribute most to")
print("accurately identifying operational anomalies?")

print("\nDataset Statistics")
print("-" * 70)
print(f"Training Samples             : {n_samples:,}")
print(f"Predictor Variables          : {n_predictors}")
print(f"Positive Events (Anomalies)  : {n_events:,}")
print(f"Negative Events (Normal)     : {n_non_events:,}")

print("\nSample Size Assessment")
print("-" * 70)
print(f"Green's Rule Minimum         : {green_rule}")
print(f"Events Per Variable (EPV)    : {epv:.2f}")

print("\nConclusion")
print("-" * 70)
print("✓ Dataset exceeds Green's Rule requirement.")
print("✓ EPV is substantially greater than the recommended minimum.")
print("✓ Dataset is statistically adequate for feature importance")
print("  analysis and supervised machine learning.")
```

## Green's Rule

Green (1991) proposed a heuristic for estimating the minimum sample size required for multivariable analyses.

The rule is expressed as:

n ≥ 50+8m

where

- **n** = minimum required sample size

- **m** = number of predictor variables

In this study:

- Predictor variables = **42**

Therefore,

n ≥ 50 + 8(42) = 386

The available training dataset contains **175,341 observations**, which is substantially larger than the minimum recommended sample size.

Although Green's Rule was originally proposed for multiple linear regression, it is included here as a supporting heuristic demonstrating that the dataset size greatly exceeds commonly accepted recommendations for multivariable analysis.

Green's Rule provides a supporting heuristic indicating that the available sample size substantially exceeds the minimum recommended for multivariable analysis. Because the primary analyses involve supervised machine learning classification, Green's Rule is complemented by the Events Per Variable (EPV) criterion, which serves as the principal sample adequacy measure for this study.

## Events Per Variable (EPV)

Because this study develops supervised machine learning classification models, Events Per Variable (EPV) provides a more appropriate measure of sample adequacy than Green's Rule.

Events Per Variable (EPV) is a widely accepted measure for assessing dataset adequacy in binary classification models.
                                   
                                   Number of Positive Events
     Events Per Variable (EPV) =  _________________________________
                                   Number of Predictor Variables
 

where

- Positive Events = observations labelled as anomalies (`label = 1`)
- Predictor Variables = independent variables used for model development

For this study:

- Positive Events = **119,341**
- Predictor Variables = **42**

Therefore,

               119341
       EPV =  ________ =2841.45
                42
       

Previous methodological research recommends a minimum EPV of **10**, while values of **20 or greater** are generally considered excellent for predictive modelling.

The calculated EPV of **2841.45** indicates that the dataset contains an exceptionally large number of anomaly observations relative to the number of predictors, providing strong evidence that the dataset is more than adequate for feature importance analysis and supervised machine learning.

Events are the number of observations belonging to the positive outcome class (label = 1), representing anomalous network flows. The EPV is computed by dividing the number of anomalous observations by the number of predictor variables used in model development.
